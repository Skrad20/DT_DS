# Восстановление золота из руды

Прототип модели машинного обучения для «Цифры». Компания разрабатывает решения для эффективной работы промышленных предприятий.

Модель должна предсказать коэффициент восстановления золота из золотосодержащей руды. Данные с параметрами добычи и очистки. 

Модель поможет оптимизировать производство, чтобы не запускать предприятие с убыточными характеристиками.

Вам нужно:

1. Подготовить данные;
2. Провести исследовательский анализ данных;
3. Построить и обучить модель.

Нужно спрогнозировать две величины:
* Эффективность обогащения чернового концентрата  - rougher.output.recovery;
* Эффективность обогащения финального концентрата - final.output.recovery.

## Описание данных

Описание данных

Технологический процесс
* Rougher feed — исходное сырье
* Rougher additions (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant
    * Xanthate **— ксантогенат (промотер, или активатор флотации);
    * Sulphate — сульфат (на данном производстве сульфид натрия);
    * Depressant — депрессант (силикат натрия).
* Rougher process (англ. «грубый процесс») — флотация
* Rougher tails — отвальные хвосты
* Float banks — флотационная установка
* Cleaner process — очистка
* Rougher Au — черновой концентрат золота
* Final Au — финальный концентрат золота

Параметры этапов
* air amount — объём воздуха
* fluid levels — уровень жидкости
* feed size — размер гранул сырья
* feed rate — скорость подачи

Наименование признаков должно быть такое:

[этап].[тип_параметра].[название_параметра]

Пример: rougher.input.feed_ag

Возможные значения для блока [этап]:
* rougher — флотация
* primary_cleaner — первичная очистка
* secondary_cleaner — вторичная очистка
* final — финальные характеристики

Возможные значения для блока [тип_параметра]:
* input — параметры сырья
* output — параметры продукта
* state — параметры, характеризующие текущее состояние этапа
* calculation — расчётные характеристики

### Расчёт эффективности обогащения

```python
def calc_recovery(
    proportion_of_gold_in_concentrate_after_flotation: float,
    proportion_of_gold_in_raw_materials_before_flotation: float,
    proportion_of_gold_in_the_tailings_after_flotation: float,
) -> float:
    """
    Возвращает расчётное значение эффективности обогащения.
    
    Эффективность обогащения рассчитывается по формуле
    recovery = ((C * (F - T)) / (F * (C-T))) * 100
    где:
    C — доля золота в концентрате после флотации/очистки;
    F — доля золота в сырье/концентрате до флотации/очистки;
    T — доля золота в отвальных хвостах после флотации/очистки.
    ----------
    Параметры:
        proportion_of_gold_in_concentrate_after_flotation: float -
            доля золота в концентрате после флотации/очистки (C).
        proportion_of_gold_in_raw_materials_before_flotation: float -
            доля золота в сырье/концентрате до флотации/очистки (F).
        proportion_of_gold_in_the_tailings_after_flotation: float -
            доля золота в отвальных хвостах после флотации/очистки (T).
    ------------
    Возращается:
        recovery: float - расчётное значение эффективности обогащения (%).
    ----------
    Выводится:
        None
    """
    return (
        (
            (
                proportion_of_gold_in_concentrate_after_flotation * (
                    proportion_of_gold_in_raw_materials_before_flotation - 
                    proportion_of_gold_in_the_tailings_after_flotation
                )
            ) / (
                proportion_of_gold_in_raw_materials_before_flotation * (
                    proportion_of_gold_in_concentrate_after_flotation -
                    proportion_of_gold_in_the_tailings_after_flotation
                )
            )
        ) * 100
    )
```

### Расчёт метрики качества

```python
def calc_sMAPE(
    target_values: pd.Series,
    predict_values: pd.Series,
) -> float:
    """
    Возвращает расчётное значение метрики качества sMAPE.
    
    Метрика рассчитывается по формуле
    sMAPE = ((1 / count_obj) * SUMM (
        (|target_values[i] - predict_values[i]|) / 
        ((||target_values[i]| + |predict_values[i]||) / 2)
        ) * 100
    где:
    target_values[i] - Значение целевого признака для объекта с 
        порядковым номером i в выборке, на которой измеряется качество.
    predict_values[i] - Значение предсказания для объекта с порядковым 
        номером i, например, в тестовой выборке.
    count_obj - Количество объектов в выборке.

    ----------
    Параметры:
        target_values: pd.Series - Значения целевого признака 
        predict_values: pd.Series - Значения предсказания признака
    ------------
    Возращается:
        sMAPE: float - расчётное значение метрики качества sMAPE (%).
    ----------
    Выводится:
        None
    
    """
    sMAPE = 1 / len(target_values) * np.sum(2*
        np.abs(target_values - predict_values) / ((np.abs(predict_values) + np.abs(target_values)))
    )
    return sMAPE * 100
```

### Удаление лишний столбцов

```python
def removing_extra_columns(
    dataset_compare : pd.DataFrame,
    dataset_delete: pd.DataFrame
) -> pd.DataFrame:
    """
    Возвращает второй датасет (dataset_delete) без столбцов,
    которых нет в первом датасете (dataset_compare). 

    ----------
    Параметры:
        dataset_compare : pd.DataFrame - датасет сравнения,
            тот по которому определяем нужен столбец или нет.
        dataset_delete: pd.DataFrame - датасет удаления,
            тот в котором удаляем лишние столбцы.
    ------------
    Возращается:
        dataset_delete: pd.DataFrame - без лишних столбцов.
    ----------
    Выводится:
        None
    """
    return dataset_delete.drop(set(dataset_delete.columns) - set(dataset_compare.columns), axis=1)
```

### Адаптер для расчёта эффективности обогащения

```python
def calc_recovery_for_apply(
    row,
    concentrate_au:str,
    feed_au: str,
    tail_au: str
):
    """
    Возращает расчётное значение эффективности обогащения.
    Адаптер для функции calc_recovery под apply датасета

    ----------
    Параметры:
        row : pd.Series - строка.
        concentrate_au: str - доля золота в концентрате после флотации/очистки;
        feed_au: str - доля золота в сырье/концентрате до флотации/очистки;
        tail_au: str - доля золота в отвальных хвостах после флотации/очистки.
    ------------
    Возращается:
        res: float - эффективность обогащения.
    ----------
    Выводится:
        None
    """
    res = calc_recovery(
        row[concentrate_au],
        row[feed_au],
        row[tail_au],
    )
    return res
```

```python
def col_substance(df_col: list, substance: str) -> set:
    """
    Выводит названия столбцов с искомым веществом.
    ----------
    Параметры:
        df_col: list - Столбцы датасета.
        substance: str - Название вещества.
    ------------
    Возращается:
        set_res: set - Названия столбцов с искомым веществом.
    ----------
    Выводится:
        None
    """
    set_res = set()
    for col in df_col:
        if substance in col:
            set_res.add(col)
    return set(set_res)
```

### Метрики регрессии

```python
def metrics_model(target_valid: pd.Series, predict: pd.Series) ->dict:
    """
    Возвращает словарь с метриками регрессии.
    ----------
    Параметры:
        target_valid: pd.Series - Истинные значения.
        predict: pd.Series - Предсказанные значения.
    -----------
    Возвращает:
        dict_res: dict - Словарь с метриками.
    """
    dict_res = {
        "MAE": mae(target_valid, predict),
        "MSE": mse(target_valid, predict),
        "RMSE": mse(target_valid, predict) ** 0.5,
        "R2": r2(target_valid, predict),
        "sMAPE": calc_sMAPE(target_valid, predict)
    }
    return dict_res
```

### Сбор метрик регрессии до подбора гиперпараметров и после

```python
def create_dataframe_estimation_model(before: dict, after:dict) -> pd.DataFrame:
    """
    Возвращает датасет с оценками модели до подбора гиперпараметров и после.
    ----------
    Параметры:
        before: dict - Словарь с оценками до подбора гиперпараметров.
        after:dict - Словарь с оценками после подбора гиперпараметров.
    -----------
    Возвращает:
        res_estimation: pd.DataFrame - Датасет с оценками модели до подбора гиперпараметров и после.
    
    """
    res_estimation = pd.DataFrame.from_dict(before, orient='index')
    res_estimation = res_estimation.rename(columns={0: "before"})
    res_estimation["after"] = after.values()
    return res_estimation
```

### Фильтрация выбросов

```python
def filter_exrand_values(df):
    for col in df.select_dtypes(include=[np.number]).columns:
        Q25 = df[col].quantile(0.07)
        Q50 = df[col].quantile(0.50)
        Q75 = df[col].quantile(0.93)
        IOR = Q75 - Q25
        top = Q50 + 1.5 * IOR
        low = Q50 - 1.5 * IOR
        df = df[(df[col] >= low) & (df[col] <= top)]
    return df.reset_index(drop=True)
```

```python
list_substances = [
    'ag',
    'pb',
    'sol',
    'au',
    'sulfate',
    'sol',
    'xanthate',
]
```

```python
list_metal = [
    'ag',
    'pb',
    'au',
]
```

```python
list_stage = [
    "rougher",
    "primary_cleaner",
    "final",
]
```

```python
dict_res = {}
dict_res_met = {}
for stage in list_stage:
    
    sum_concentrate_input = pd.Series(np.zeros(len(df_full_not_tail)))
    sum_concentrate_output = pd.Series(np.zeros(len(df_full_not_tail)))
    
    sum_concentrate_input_met = pd.Series(np.zeros(len(df_full_not_tail)))
    sum_concentrate_output_met = pd.Series(np.zeros(len(df_full_not_tail)))
    for sub in list_substances:
        
        set_col_sub = col_substance(tree_level, sub)
        for col_sub in set_col_sub:
            try:
                df_temp = df_full_not_tail.loc[:, (stage, 'input', col_sub)]
                sum_concentrate_input += df_temp
                if sub in list_metal:
                    sum_concentrate_input_met += df_temp
            except:
                pass
            try:
                df_temp = df_full_not_tail.loc[:, (stage, 'output', col_sub)]
                sum_concentrate_output += df_temp
                if sub in list_metal:
                    sum_concentrate_output_met += df_temp
            except:
                pass
    dict_res[stage + ':input'] = sum_concentrate_input
    dict_res[stage + ':output'] = sum_concentrate_output
    dict_res_met[stage + ':input'] = sum_concentrate_input_met
    dict_res_met[stage + ':output'] = sum_concentrate_output_met

```

```python
sum_conc_all = pd.DataFrame.from_dict(dict_res).astype('float')
sum_conc_all.describe()
```

```python
sum_conc_met = pd.DataFrame.from_dict(dict_res_met).astype('float')
sum_conc_met.describe()
```
