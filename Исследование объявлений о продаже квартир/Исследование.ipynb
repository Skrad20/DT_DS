## Функции и их тестирование

### Функция категоризации населенных пунктов

```python
def categorization_localities(row, col: str):
    """
    Функция создает столбец в датасете с категориями населенных пунктов
    Выделим три основные категории: деревня, поселок, город
    
    """
    locality_name = row[col]
    
    if pd.isna(locality_name):
        row['category_locality_name'] = 'неизвестно'
        return row

    locality_name = str(row[col]).lower()
    if 'деревня' in locality_name or 'дерев' in locality_name:
        row['category_locality_name'] = 'деревня'
        return row
    elif  'посёлок' in locality_name or 'поселок' in locality_name:
        row['category_locality_name'] = 'поселок'
        return row
    elif  'санкт-петербург' in locality_name:
        row['category_locality_name'] = 'Санкт-Петербург'
        return row
    elif  'село' in locality_name:
        row['category_locality_name'] = 'село'
        return row
    elif  'товарищество' in locality_name or 'садовое товарищество' in locality_name:
        row['category_locality_name'] = 'товарищество'
        return row
    elif len(locality_name.split()) == 1:
        row['category_locality_name'] = 'город'
        return row
    else:
        row['category_locality_name'] = 'деревня'
        return row
```

Протестируем функцию. Создадим тестовые данные с известными ответами

### Связь пропусков

Проверим, есть ли связь в пропусках данных

Выдели датасет для анализа связи

```python
df_analis = df[[
    'ceiling_height', 'floors_total', 'living_area',
    'is_apartment','kitchen_area', 'balcony',
    'locality_name', 'airports_nearest', 'cityCenters_nearest',
    'parks_around3000', 'parks_nearest', 'ponds_around3000',
    'ponds_nearest', 'days_exposition'
]]
```

Проведем замены значение на значения "есть" и "нет". 
Создадим правила пометок наличия и отсутствия значений

```python
availability_selection_criteria = ~df_analis.isna()
absence_selection_criteria = df_analis.isna()
```

Пометим в новом датасете заначения

```python
df_analis = df_analis.where(availability_selection_criteria, 'нет')
df_analis = df_analis.where(absence_selection_criteria, 'есть')
```

Создадим кросс-таблицы для все параметров с пропусками. 

Нас интересует общее количество пропусков и процент общих пропусков от всех пропусков в двух параметрах. Это позволит оценить насколько связаны пропуски в двух параметрах. 

```python
df_res_cross = pd.DataFrame(columns=[
    'first_par', 'second_par', 'all_nan','percentage_linked_nan'
])
count = 0
for col_first in df_analis.columns:
    for col_second in df_analis.columns:
        if col_first != col_second:
            cross = pd.crosstab(
                df_analis[col_first],
                df_analis[col_second],
            )
            list_sort = sorted([col_first, col_second])
            df_res_cross.loc[count, 'first_par'] = list_sort[0]
            df_res_cross.loc[count, 'second_par'] = list_sort[1]
            df_res_cross.loc[count, 'all_nan'] = (
                cross.sum().sum() - cross.iloc[0, 0]
            )
            df_res_cross.loc[count, 'percentage_linked_nan'] =(
                (cross.iloc[1, 1] / (cross.sum().sum() - cross.iloc[0, 0])) * 100
            )
            count += 1
```

Удалим получившиеся дубликаты. Отсортируем по проценту общих пропусков. И оставим те записи, в которых процент общих пропусков более 50%

```python
df_res_cross = (
    df_res_cross
    .sort_values(
        by='percentage_linked_nan',
        ascending=False)
    .drop_duplicates(
        subset=['first_par','second_par']
    )
    .query("percentage_linked_nan > 50")
    .reset_index(drop=True)
)
```
